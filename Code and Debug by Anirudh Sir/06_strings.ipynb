```python
# 1. https://leetcode.com/problems/remove-outermost-parentheses/description/

s = "(()())(())"

count = 0 
result = []
for i in s:
    if i == "(":
        count += 1
    else:
        count -= 1
    if i == "(" and count > 1: #exclude the first open bracket i.e. count==1
        result.append(i)
    elif i == ")" and count > 0: #exclue the last closing bracket i.e. count==0
        result.append(i)
        
"".join(result)       
```

```python
# 4. Find the longest common prefix

def longest_common_prefix(strs):
    if not strs:
        return ""
    
    for i in range(len(strs[0])):
        current_char = strs[0][i]
        for word in strs[1:]:
            if len(word) < i or word[i] != current_char:
                return strs[0][:i]
    
    return strs[0]

strs = ["flower", "flow", "flight"]
print(longest_common_prefix(strs))  # Output: "fl"
```

```python
# 5. https://leetcode.com/problems/isomorphic-strings/description/

def isIsomorphic(s, t):
    if len(s) != len(t):
        return False
    s_to_t_map = {}
    t_to_s_map = {}
    for i in range(len(s)):
        if s[i] in s_to_t_map:
            if s_to_t_map[s[i]] != t[i]:
                return False
        else:
            s_to_t_map[s[i]] = t[i]
            
        if t[i] in t_to_s_map:
            if t_to_s_map[t[i]] != s[i]:
                return False
        else:
            t_to_s_map[t[i]] = s[i]
            
    return True


print(isIsomorphic(s = "egg", t = "add")) #True
print(isIsomorphic(s = "foo", t = "bar")) #False
print(isIsomorphic(s = "paper", t = "title")) #True
print(isIsomorphic(s = "badc", t = "baba")) #False
```

```python
# 7. Check if the two strings are anagram

def isAnagram(s, t):
    if len(s) != len(t):
        return False

    freq_map = {}
    for char in s:
        freq_map[char] = freq_map.get(char, 0) + 1

    for char in t:
        if char not in freq_map or freq_map[char] == 0: #char doesn't exist in the previous string or occuring more than previous string
            return False
        freq_map[char] -= 1

    return True

print(isAnagram("anagram", "nagaram"))  # True
print(isAnagram("rat", "car"))          # False
print(isAnagram("aabb", "aabc"))        # False
print(isAnagram("", ""))                # True
print(isAnagram("a", "a"))              # True
```

```python
# 8. https://leetcode.com/problems/sort-characters-by-frequency/

def frequencySort(s):
    f_map = {}
    for char in s:
        f_map[char] = f_map.get(char, 0) + 1

    char_list = list(f_map.items())
    char_list.sort(key=lambda x: x[1], reverse=True)

    return "".join(char * freq for char, freq in char_list)
    
print(frequencySort("tree"))
print(frequencySort("cccaaa"))
print(frequencySort("Aabb"))
```

```python
# 9. https://leetcode.com/problems/maximum-nesting-depth-of-the-parentheses/description/

def maxDepth(s):
    depth = 0
    max_depth = 0
    for char in s:
        if char == "(":
            depth += 1
            max_depth = max(max_depth, depth)
        elif char == ")":
            depth -= 1
    return max_depth

print(maxDepth("(1+(2*3)+((8)/4))+1"))       # Output: 3
print(maxDepth("(1)+((2))+(((3)))"))         # Output: 3
print(maxDepth("()(())((()()))"))            # Output: 3
print(maxDepth("()"))                        # Output: 1
print(maxDepth("(((())))"))                  # Output: 4
print(maxDepth("no(parens)here"))            # Output: 1
```

```python
#11. https://leetcode.com/problems/minimum-window-substring/description/

from collections import Counter 

def minWindow(s: str, t: str) -> str:
    if len(t) > len(s):
        return ""
    
    freq_map = Counter(t)
    total = len(t)
    start = 0
    smallest = ""
    min_len = float('inf')

    for end in range(len(s)):
        char = s[end]
        if char in freq_map:
            freq_map[char] -= 1
            if freq_map[char] >= 0:    #abhi sare chars mile nahi hain, isliye count 0 ya usse jyada hai toh total mein se minus kro
                total -= 1             #lekin agr ye value -ve hoti mtlb t mein jitne the wo pehle hi minus ho chuke hain ab extra char se total ko fark
                                       #nahi padega
        while total == 0:
            # update smallest window
            window_len = end - start + 1
            if window_len < min_len:
                min_len = window_len
                smallest = s[start:end+1]

            # shrink from left
            left_char = s[start]
            if left_char in freq_map:
                freq_map[left_char] += 1
                if freq_map[left_char] > 0:  #agr ye value 0 ya kum hoti mtlb wo extra char hai, its not a part of t, toh total mein add krne ka mtlb nahi
                    total += 1
            start += 1

    return smallest

# ✅ Test
print(minWindow("ADOBECODEBANC", "ABC"))  # Output: "BANC"
print(minWindow("a", "aa"))               # Output: ""
print(minWindow("aaflslflsldkalskaaa", "aaa"))  # Output: "aaa"
```

```python
#12. https://leetcode.com/problems/valid-palindrome-ii/description/

def is_palindrome(string):
    """Check if the given string is a palindrome."""
    return string == string[::-1]

def check_substring_palindrome(input_string):
    """Check if a string can be a palindrome by removing one character."""
    start, end = 0, len(input_string) - 1
    
    while start <= end:
        if input_string[start] != input_string[end]:
            # Check if removing either the start or the end character results in a palindrome
            return is_palindrome(input_string[start + 1: end + 1]) or is_palindrome(input_string[start: end])
        start += 1
        end -= 1
    return True

# Example usage
input_string = "abc"
result = check_substring_palindrome(input_string)
print(result)
```

```python
#13. https://leetcode.com/problems/group-anagrams/description/
# This can be done in O(m x n) in next cell using freq_map


from collections import defaultdict

def group_anagrams(words):
    """Group words that are anagrams of each other."""
    anagrams = defaultdict(list)
    
    for word in words:
        # Sort characters in the word to form a key
        key = ''.join(sorted(word))
        anagrams[key].append(word)
    
    # Return grouped anagrams as a list of lists
    return list(anagrams.values())

# Example usage
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
result = group_anagrams(words)
print(result)

# The time complexity:
#     Sorting each string → O(m log m) (where m is string length).
#     For n strings → O(n × m log m) overall.
```

```python
#13. https://leetcode.com/problems/group-anagrams/description/

def group_anagrams_optimized(words):
    from collections import defaultdict
    anagrams = defaultdict(list)
    
    for word in words:
        # Count frequency of each character
        count = [0] * 26
        for char in word:
            count[ord(char) - ord('a')] += 1
        anagrams[tuple(count)].append(word)
    
    return list(anagrams.values())

# Example usage
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
result = group_anagrams(words)
print(result)
```

```python
# 14. https://leetcode.com/problems/longest-palindromic-substring/description/

def longestPalindrome(s):
    if not s:
        return ""
    
    n = len(s)
    result = ""
    
    def expand(left, right):
        nonlocal result
        # expand while valid and characters match
        while left >= 0 and right < n and s[left] == s[right]:
            cur_len = right - left + 1
            if cur_len > len(result):
                result = s[left:right+1]
            left -= 1
            right += 1
    
    for i in range(n):
        # odd length palindromes (single center)
        expand(i, i)
        # even length palindromes (center between i and i+1)
        expand(i, i + 1)
    return result

print(longestPalindrome("babad"))
print(longestPalindrome("cbbd"))
print(longestPalindrome("abacdgh"))
```

```python
 # 15. https://leetcode.com/problems/longest-substring-with-at-most-k-distinct-characters/

def longest_substring_k_distinct(s, k):
    if not s or k == 0:
        return 0

    start = 0
    freq_map = {}
    longest = 0

    for end, char in enumerate(s):
        freq_map[char] = freq_map.get(char, 0) + 1

        # If there are more than k distinct chars, shrink the window
        while len(freq_map) > k:
            freq_map[s[start]] -= 1
            if freq_map[s[start]] == 0:
                del freq_map[s[start]]
            start += 1

        # Update the longest valid window
        longest = max(longest, end - start + 1)

    return longest


# ✅ Example test
s = "abbdbbbc"
k = 2
print(longest_substring_k_distinct(s, k))  # Output: 5 ("bbbdb" or "dbbbb")       
```

```python
# 16. https://leetcode.com/problems/first-unique-character-in-a-string/description/

from collections import Counter

def first_unique_char_index(s):
    # First pass: count occurrences of each character
    freq = Counter(s)

    # Second pass: find the first character with count == 1
    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1  # if all characters repeat


# ✅ Example tests
print(first_unique_char_index("abcdab"))   # 2 ('c')
print(first_unique_char_index("aabb"))     # -1 (no unique)
print(first_unique_char_index("leetcode")) # 0 ('l')
print(first_unique_char_index(""))         # -1 (empty string)
```
