```python
# 1. https://leetcode.com/problems/assign-cookies/description/

def findContentChildren(greed_factors, cookie_sizes):
    # Sort greed factors and cookie sizes in ascending order
    greed_factors.sort()
    cookie_sizes.sort()

    child_idx = 0
    cookie_idx = 0
    content_children = 0

    # Try to satisfy each child with the smallest possible cookie
    while child_idx < len(greed_factors) and cookie_idx < len(cookie_sizes):
        if greed_factors[child_idx] <= cookie_sizes[cookie_idx]:
            content_children += 1
            child_idx += 1  # Move to the next child (this one is content)
        cookie_idx += 1  # Try the next cookie

    return content_children
        
print(findContentChildren([1,2,3], [1,1])) # Output: 1
print(findContentChildren([1,2], [1,2,3])) # Output: 2
```

```python
# 2. https://www.geeksforgeeks.org/problems/fractional-knapsack-1587115620/1

def fractionalKnapsack(val, wt, capacity):
    items = list(zip(val, wt))

    # sort the items by value, weight ratio
    items.sort(key = lambda x: x[0] // x[1], reverse = True)
    final_value = 0
    rem_cap = capacity
    
    # Keep adding the item till we have capacity to accomodate whole item
    for val, wt in items:
        if rem_cap - wt > 0:
            rem_cap -= wt
            final_value += val
        # Add the fraction of item according to remaining capacity
        else:
            final_value += rem_cap * (val/wt)
            break
    return final_value

print(fractionalKnapsack([60, 100, 120], [10, 20, 30], 50)) # Output = 240
print(fractionalKnapsack([500], [30], 10)) # Output = 166.66
```

```python
# 3. https://www.geeksforgeeks.org/problems/-minimum-number-of-coins4426/1

def findMinCoins(value):
    den = [ 1, 2, 5, 10, 20, 50, 100, 200 ]
    n = len(den)
    i = n-1

    # find the just smaller note than value
    while den[i] > value:
        i -= 1
        
    total = den[i]
    count = 1
    
    while value > total:
        if total + den[i] > value:   # if using this note is too big to use than move to smaller one
            i -= 1
        else:
            total += den[i]
            count += 1
    return count

print(findMinCoins(39))
print(findMinCoins(121))
```

```python
# 4. https://www.geeksforgeeks.org/problems/n-meetings-in-one-room-1587115620/1

def maxMeetings(start, end):
    n = len(start)
    # we can store (end, start) or (start, end) but we want to sort by end time
    meetings = list(zip(end, start))
    
    # Sort by end time (first element of tuple)
    meetings.sort()
    # Now meetings are ordered so that the earliest finishing meeting comes first

    # We always take the first meeting in this sorted list
    count = 1
    last_end_time = meetings[0][0]  # the end time of the first selected meeting

    # Now iterate through remaining meetings
    for i in range(1, n):
        meeting_end, meeting_start = meetings[i]  # note: stored as (end, start)
        # We can attend this meeting if it starts **strictly after** last meeting ended
        if meeting_start > last_end_time:
            count += 1
            last_end_time = meeting_end
    
    return count


# Test examples
print(maxMeetings([1, 3, 0, 5, 8, 5], [2, 4, 6, 7, 9, 9]))  # expect 4
print(maxMeetings([10, 12, 20], [20, 25, 30]))  # expect 1
```

```python
# 5. https://leetcode.com/problems/jump-game/description/

def canJump(steps):
    n = len(steps)
    max_index = 0
    for i in range(n):
        if i > max_index:
            return False
        max_index = max(max_index, i+steps[i])
    return True

print(canJump([2,3,1,1,4])) #output: True
print(canJump([3,2,1,0,4])) #output: False
```

```python
# 6. https://leetcode.com/problems/jump-game-ii/description/

def minJump(nums):
    n = len(nums)
    jumps = 0
    curr_end = 0
    farthest = 0

    for i in range(n - 1):  # we don't need to jump from the last index
        farthest = max(farthest, i + nums[i])
        
        # End of the current jump range
        if i == curr_end:
            jumps += 1
            curr_end = farthest

    return jumps

minJump([2,3,1,1,4])
```

```python
def minJump(jumps):
    n = len(jumps)
    min_jump = 0
    left = 0
    right = 0
    while right < n-1:
        farthest = 0
        for i in range(left, right+1):
            farthest = max(farthest, i + jumps[i])
        left = right + 1
        right = farthest
        min_jump += 1
    return min_jump
        
minJump([2,3,1,1,4])
```

```python
# https://www.geeksforgeeks.org/problems/minimum-platforms-1587115620/1
# not the optimal, on my own

def minPlatforms(arr, dep):
    n = len(arr)
    schedule = list(zip(dep, arr))
    schedule.sort()
    platforms = 1
    occupied_till = {1: schedule[0][0]}
    for i in range(1, n):
        dep_time = schedule[i][0]
        arr_time = schedule[i][1]
        for key, value in occupied_till.items():
            if arr_time > value:
                occupied_till[key] = dep_time
                break
        else:
            platforms += 1
            occupied_till[platforms] = dep_time

    return platforms
            
print(minPlatforms(arr = [900, 940, 950, 1100, 1500, 1800], dep = [910, 1200, 1120, 1130, 1900, 2000]))
print(minPlatforms(arr = [900, 1235, 1100], dep = [1000, 1240, 1200]))
print(minPlatforms(arr = [1000, 935, 1100], dep = [1200, 1240, 1130]))
```

```python
# 7. https://www.geeksforgeeks.org/problems/minimum-platforms-1587115620/1

def minPlatforms(arr, dep):
    n = len(arr)
    arr.sort()
    dep.sort()

    platform_needed = 1
    max_platforms = 1

    i = 1  # pointer for arrival
    j = 0  # pointer for departure

    while i < n and j < n:
        if arr[i] <= dep[j]:
            # New train arrives before the earliest one departs → need new platform
            platform_needed += 1
            i += 1
        else:
            # Train departs before next one arrives → free one platform
            platform_needed -= 1
            j += 1

        max_platforms = max(max_platforms, platform_needed)

    return max_platforms

print(minPlatforms(arr = [900, 940, 950, 1100, 1500, 1800], dep = [910, 1200, 1120, 1130, 1900, 2000]))
print(minPlatforms(arr = [900, 1235, 1100], dep = [1000, 1240, 1200]))
print(minPlatforms(arr = [1000, 935, 1100], dep = [1200, 1240, 1130]))
```
