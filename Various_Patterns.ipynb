```python
# 1. https://leetcode.com/problems/top-k-frequent-elements/

def topKFrequent(nums, k):
    freq_map = {}
    for n in nums:
        freq_map[n] = freq_map.get(n, 0) + 1   
    tuples = sorted(freq_map.items(), key=lambda item: item[1], reverse = True)[:k]
    return list([key for key,value in tuples])
        
print(topKFrequent([1,1,1,2,2,3], 2)) #output = [1,2]
```

```python
# 2. https://leetcode.com/problems/product-of-array-except-self/

def productExceptSelf(nums):
    n = len(nums)
    prefix_product = [1] * n
    for i in range(1, n-1):
        prefix_product[i+1] = prefix_product[i] * nums[i]
    print(prefix_product)

    suffix = 1
    for i in range(n-2,-1,-1):
        suffix = suffix * nums[i+1]
        prefix_product[i] = prefix_product[i] * suffix
    return prefix_product


print(productExceptSelf(nums = [1,2,3,4]))
```

```python
# 5. Binary search in the rotated sorted array

arr = [4, 5, 6, 7, 8, 9, 1, 2, 3]
target = 1

def find_num(arr, target):
    start=0
    end=len(arr)-1
    while start<=end:
        mid=(start+end)//2
        if arr[mid] == target:
            return mid
        elif arr[start] <= arr[mid]:
            print("left array is sorted")
            if arr[start] <= target <= arr[mid]:
                end = mid-1
            else:
                start = mid+1
        else:
            print("Right array is sorted")
            if arr[mid] <= target <= arr[end]:
                start = mid+1
            else:
                end = mid-1
    return -1

ans = find_num(arr, target)
print(ans)
```

```python
# 6. https://leetcode.com/problems/subarray-sum-equals-k/description/

def subarraySum(nums, k):
    # Initialize the dictionary to store the frequency of prefix sums.
    prefix_sum_count = {0: 1}  # We assume a sum of 0 has occurred once (before the array starts)
    current_sum = 0
    count = 0
    
    # Iterate through the array
    for num in nums:
        # Update the current cumulative sum
        current_sum += num
        
        # Check if current_sum - k has been seen before.
        if current_sum - k in prefix_sum_count:
            count += prefix_sum_count[current_sum - k]
        
        # Update the dictionary with the current cumulative sum.
        prefix_sum_count[current_sum] = prefix_sum_count.get(current_sum, 0) + 1
    
    return count
    
                  
print(subarraySum(nums = [0,1,1,1,1], k = 0)) #Output: 1
print(subarraySum(nums = [1,2,3], k = 3)) #Output: 2
print(subarraySum(nums = [9,4,0,20,3,10,5], k = 33)) #Output: 3
```

```python
# 7. Find all the sub sequences of an array
nums = [5,9,7]

def get_all(nums):
    result = []
    
    def solve(index, subset):
        if index >= len(nums):
            result.append(subset.copy())
            return
        # Include current element
        subset.append(nums[index])
        solve(index + 1, subset)
        # Exclude current element
        subset.pop()
        solve(index + 1, subset)

    solve(0, [])
    return result

# Test
print(get_all([5, 9, 7]))
```

```python
# 8. Selection sort: Find the mininum and swap the value with ith index

def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if nums[j] < nums[min_index]:
                min_index = j
        nums[i], nums[min_index] = nums[min_index], nums[i]
    return nums

selection_sort([1,4,2,6,3,8,5,7])
```

```python
# 9. https://leetcode.com/problems/valid-parentheses/

def isValid(s: str) -> bool:
    bracket_map = {
    ")": "(",
    "]": "[",
    "}": "{"
    }
    stack = []
    for i in s:
        if i in bracket_map.values(): #open mile toh stack mein dalo
            stack.append(i)
        else:
            if not stack:  #jab closing bracket mile tab stack empty nahi hona chahiye, iska mtlb hai ki bina open k hi direct closing mil gaya
                return False
            if bracket_map[i] != stack[-1]: #jab closing bracket mile toh uska jodidar stack ka last element hona chahiye
                return False
            else:                         # valid closing mile toh uske jodidar yani ki last element in stack ko bahar nikal do
                stack.pop()
    return not stack                      # agar stack empty hai mtlb jitne open mile utne hi closing mile wo bhi sahi order mein toh return True

print(isValid(s = "()")) #True
print(isValid(s = "()[]{}")) #True
print(isValid(s = "(]")) #False
print(isValid(s = "([])")) #True
print(isValid(s = "([)]")) #False
```

```python
# 10. https://leetcode.com/problems/assign-cookies/description/

def findContentChildren(greed_factors, cookie_sizes):
    # Sort greed factors and cookie sizes in ascending order
    greed_factors.sort()
    cookie_sizes.sort()

    child_idx = 0
    cookie_idx = 0
    content_children = 0

    # Try to satisfy each child with the smallest possible cookie
    while child_idx < len(greed_factors) and cookie_idx < len(cookie_sizes):
        if greed_factors[child_idx] <= cookie_sizes[cookie_idx]:
            content_children += 1
            child_idx += 1  # Move to the next child (this one is content)
        cookie_idx += 1  # Try the next cookie

    return content_children
        
print(findContentChildren([1,2,3], [1,1])) # Output: 1
print(findContentChildren([1,2], [1,2,3])) # Output: 2
```

```python
# 11. Rotate array to right by k places

def rotate_array_by_k(arr, k):
    def reverse_array(tmp_arr, left, right):
        i,j = left,right
        while i < j:
            tmp_arr[i], tmp_arr[j] = tmp_arr[j], tmp_arr[i]
            i += 1
            j -= 1
        return tmp_arr

    n = len(arr)
    reverse_array(arr, 0, n-k-1)
    reverse_array(arr, n-k, n-1)
    reverse_array(arr, 0, n-1)
    return arr

arr = [1,2,3,4,5,6,7]
rotate_array_by_k(arr, 2)
```

```python
# 12. First and last occurence of a number in an array

def first_and_last(nums, target):
    def lower_bound(nums, target):
        lb = -1
        low, high = 0, len(nums)-1
        while low <= high:
            mid = (low+high) // 2
            if nums[mid] >= target:
                if nums[mid] == target:
                    lb = mid
                high = mid-1
            else:
                low = mid+1
        return lb

    def upper_bound(nums, target):
        ub = -1
        low, high = 0, len(nums)-1
        while low <= high:
            mid = (low+high) // 2
            if nums[mid] > target:
                high = mid-1
            else:
                if nums[mid] == target:
                    ub = mid
                low = mid+1
        return ub

    return [lower_bound(nums, target), upper_bound(nums, target)]

print(first_and_last(nums = [1,2,3,3,4,5,6], target = 3))   
```
