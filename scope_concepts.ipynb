```python
# üß© 2Ô∏è‚É£Ô∏è‚É£ Accessing with self.x vs Test.x vs x

class Demo:
    x = 100
    def show(self):
        x = 200
        print(x)
        print(self.x) # this prints 100 because, python will look if self.x is defined anywhere, if it doesn't find then it start looking on class level 
        print(Demo.x)
        
d = Demo()
d.show()

# Question:
# 	‚Ä¢ Explain the output.
# 	‚Ä¢ Which x is local, instance, and class variable?
# üëâ Checks LEGB rule (Local ‚Üí Enclosing ‚Üí Global ‚Üí Built-in) and how self works.
```

```python
# üß© 3Ô∏è‚É£Ô∏è‚É£ Modifying class variable via instance

class Demo:
    count = 0
    def __init__(self):
        self.count += 1
        
d1 = Demo()
d2 = Demo()

print(Demo.count)
print(d1.count)

# Question:
# Why is Demo.count still 0? - Reasoning is given below in the screenshot
# üëâ Tests your understanding that self.count += 1 creates a new instance variable, not modify the class one.
```

```python
# üß© 4Ô∏è‚É£Ô∏è‚É£ Using cls.x vs self.x

class Sample:
    x = 5
    @classmethod
    def set_value(cls, val):
        cls.x = val
    def show(self):
            print(self.x)
        
s1 = Sample()
s2 = Sample()

Sample.set_value(100)

s1.show()
s2.show()

# Question:
# Explain how cls.x changes for all instances.
# üëâ Checks @classmethod behavior and shared class state.
```

```python
# üß© 7Ô∏è‚É£Ô∏è‚É£ Static method variable access

class Test:
    x = 10
    @staticmethod
    def show():
        print(Test.x)
        
Test.show()

# Question:
# Why can a static method access Test.x but not self.x? - self parameter hai kaha show() mein, 
#                                                         aur hota bhi toh show call krte time obj bhi toh pass krna padta
# üëâ Tests difference between @staticmethod (no self/cls) vs others.
```

```python
# üß© 8Ô∏è‚É£Ô∏è‚É£ Difference between instance, class, and static methods

class Example:
    data = 50
    def instance_m(self):
        print("Instance method", self.data)
    @classmethod
    def class_m(cls):
        print("Class method", cls.data)
    @staticmethod
    def static_m():
        print("Static method", Example.data)
            
obj = Example()

obj.instance_m() # self.data kahi define nahi kiya toh class level se le aya
obj.class_m()    # iska toh kaam hi hai class variables access krna
obj.static_m()   # self aur cls toh isme chalega nahi toh classname se access kr liya

# Question:
# Explain outputs and difference in what each method can access.
```
