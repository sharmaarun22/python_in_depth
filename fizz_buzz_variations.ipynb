```python
# üü¢ 1Ô∏è‚É£ Basic FizzBuzz

for i in range(9, 16):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# üü° 2Ô∏è‚É£ Dynamic Divisors (user-defined)

fizz = int(input("Enter fizz divisor: "))
buzz = int(input("Enter buzz divisor: "))
n = int(input("Enter upper limit: "))

for i in range(1, n+1):
    out = ""
    if i % fizz == 0:
        out += "Fizz"
    if i % buzz == 0:
        out += "Buzz"
    print(out or i)
```

```python
# üü† 3Ô∏è‚É£ FizzBuzz without % (modulo)

fizz, buzz = 3, 5
fizz_count, buzz_count = 1, 1

for i in range(9, 16):
    fizz_count += 1
    buzz_count += 1
    out = ""
    if fizz_count == fizz:
        out += "Fizz"
        fizz_count = 0
    if buzz_count == buzz:
        out += "Buzz"
        buzz_count = 0
    print(out or i)
```

```python
# üü£ 4Ô∏è‚É£ Generalized FizzBuzz (dict-based rules)

rules = {3: "Fizz", 5: "Buzz", 7: "Bazz"}

for i in range(1, 106):
    out = "".join(word for num, word in rules.items() if i % num == 0)
    print(out or i)
```

```python
# üîµ 5Ô∏è‚É£ One-liner FizzBuzz

print(['Fizz'*(i%3==0) + 'Buzz'*(i%5==0) or i for i in range(1, 101)])
```

```python
# üü§ 6Ô∏è‚É£ Contains the digit rule

# üëâ ‚ÄúFizz‚Äù if divisible by 3 or contains 3.
# üëâ ‚ÄúBuzz‚Äù if divisible by 5 or contains 5.

for i in range(1, 51):
    out = ""
    if i % 3 == 0 or "3" in str(i):
        out += "Fizz"
    if i % 5 == 0 or "5" in str(i):
        out += "Buzz"
    print(out or i)
```

```python
# üî¥ 7Ô∏è‚É£ Recursive FizzBuzz (no loops)

def fizzbuzz(n, i=1):
    if i > n:
        return
    out = ""
    if i % 3 == 0:
        out += "Fizz"
    if i % 5 == 0:
        out += "Buzz"
    print(out or i)
    fizzbuzz(n, i+1)

fizzbuzz(20)
```

```python
# ‚ö´ 8Ô∏è‚É£ FizzBuzz without if or else

for i in range(1, 101):
    print({(0, 0): "FizzBuzz", (0, 1): "Fizz", (1, 0): "Buzz"}.get((i%3, i%5), i))
```

```python
# üß† 9Ô∏è‚É£ Config-driven FizzBuzz (from JSON/dict)

import json

config = '{"rules": {"3": "Fizz", "5": "Buzz", "7": "Bazz"}, "limit": 30}'
data = json.loads(config)

for i in range(1, data["limit"] + 1):
    out = "".join(v for k, v in data["rules"].items() if i % int(k) == 0)
    print(out or i)
```

```python
# üß© üîü Chain of Responsibility (OOP FizzBuzz)

class Rule:
    def apply(self, n): return ""

class FizzRule(Rule):
    def apply(self, n): return "Fizz" if n % 3 == 0 else ""

class BuzzRule(Rule):
    def apply(self, n): return "Buzz" if n % 5 == 0 else ""

rules = [FizzRule(), BuzzRule()]

for i in range(1, 21):
    out = "".join(rule.apply(i) for rule in rules)
    print(out or i)
```

```python
# ‚öôÔ∏è 1Ô∏è‚É£1Ô∏è‚É£ Data-driven / Scalable FizzBuzz

# This one reads divisors and words dynamically (handy for test automation):

rules = input("Enter rules like 3:Fizz,5:Buzz,7:Bazz: ")
rule_dict = dict(item.split(":") for item in rules.split(","))

for i in range(1, 31):
    out = "".join(v for k, v in rule_dict.items() if i % int(k) == 0)
    print(out or i)
```

```python
# üßµ 1Ô∏è‚É£2Ô∏è‚É£ Threaded (Concurrent) FizzBuzz

# üëâ Great for advanced interviews:

import threading

n = 20
lock = threading.Lock()
current = 1

def fizz():
    global current
    while current <= n:
        with lock:
            if current % 3 == 0 and current % 5 != 0:
                print("Fizz")
                current += 1

def buzz():
    global current
    while current <= n:
        with lock:
            if current % 5 == 0 and current % 3 != 0:
                print("Buzz")
                current += 1

def fizzbuzz():
    global current
    while current <= n:
        with lock:
            if current % 15 == 0:
                print("FizzBuzz")
                current += 1

def number():
    global current
    while current <= n:
        with lock:
            if current % 3 != 0 and current % 5 != 0:
                print(current)
                current += 1

threads = [threading.Thread(target=f) for f in (fizz, buzz, fizzbuzz, number)]
for t in threads: t.start()
for t in threads: t.join()
```
